```python
import numpy as np
import pandas as pd
import pylab
import scipy.stats
df = pd.read_csv("data.csv")
pylab.style.use("ggplot")
```

```python
def get_options(name):
    return list(df[name].drop_duplicates())
```

```python
sorted(get_options("N"))
```

```python
def filter(**args):
    res = pd.DataFrame(df)
    for i in args.keys():
        res = res[res[i]==args[i]]
    return res
```

```python
print(get_options("N"))
print(get_options("r"))
print(get_options("len"))
```

```python
frame = filter(len=30)
pylab.scatter(frame["Call Counts"],frame["Time"],alpha=0.5)
scipy.stats.linregress(frame["Call Counts"],frame["Time"])
```

```python
frame = filter(r=0.17, len=15)
pylab.scatter(frame["N"],frame["Time"],alpha=0.5)
```
